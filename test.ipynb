```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# 1. 품사 분석기 준비 (Mecab: pip install python-mecab-ko 필요)
from mecab import MeCab
mecab = MeCab()

# 2. 한글 문장을 품사 시퀀스로 변환하는 함수
def to_pos_sequence(sentence):
    return " ".join([tag for _, tag in mecab.pos(sentence)])

# 3. 문장 리스트 준비 (예시)
sentence_list = [
    "나는 학교에 갔어요.",
    "나는 오늘 학교에 다녀왔습니다.",
    "너는 동물원에 갔구나.",
    "학교에서 밥을 먹었다.",
    "나는 피자를 먹었다.",
    "아침에 일어나서 준비했다.",
    "학교에 친구가 많다.",
    "나는 집에 돌아왔다.",
    "오늘 날씨가 정말 좋다.",
    "동물원에 가서 코끼리를 봤다."
]
target_sentence = "나는 학교에 갔어요."

# 4. 전체 문장 품사 시퀀스 변환
sentence_list_pos = [to_pos_sequence(s) for s in sentence_list]
target_pos = to_pos_sequence(target_sentence)

# 5. 벡터화 + 코사인 유사도 계산
vectorizer = CountVectorizer()
X = vectorizer.fit_transform([target_pos] + sentence_list_pos)
sims = cosine_similarity(X[0], X[1:])[0]

# 6. Top N개 유사 문장 출력
top_n = 3
top_idxs = sims.argsort()[::-1][:top_n]
for rank, idx in enumerate(top_idxs, start=1):
    print(f"{rank}. 문장: '{sentence_list[idx]}' | 구조 유사도: {sims[idx]:.3f}")

# 필요시 품사 시퀀스도 같이 출력
# print(sentence_list_pos[idx])
```
